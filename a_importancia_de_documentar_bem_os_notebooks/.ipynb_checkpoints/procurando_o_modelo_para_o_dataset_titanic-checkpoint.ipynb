# 0. Sobre o notebook 

**Problema principal: classificar quem sobreviveu ao naufrágio ou não.**


Objevitos: 
- Estudar o dados referentes ao dataset do titanic(veja neste [link](https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv)) que irão para o modelo;
- Verificar a possibilidade de tratamento dos dados;
- Criar uma pipeline de criação do **modelo de classificação** com: busca pelo melhor conjunto de hiperparâmetros e treinamento do modelo;

#### O que se trata esse dataset?

Resumindo que há na descrição do [Kaggle](https://www.kaggle.com/c/titanic/overview):

Durante o trágico evento que envolveu o "inafundável” O RMS Titanic afundou após colidir com um iceberg, causando 1502 mortes de um total de 2224 passageiros. 

Enquanto isso, foi notado certos padrões relacionados aos sobreviventes do titanic, percebendo que estes padrões poderia ser a causa do porquê destas pessoas sobreviverem.

#### Qual é o significado de cada coluna deste dataset?

1. `PassengerId`(int) = indetificação do passageiro no navio. ID no dataset;
2. `Survived`(int) = 1 ou 0, indicando se sobreviveu ou não naufrágio, respectivamente;
3. `Pclass`(int) = 1,2 ou 3 indicando o tipo de classe de passagem hospedada no navio. Quanto menor melhor;
4. `Name`(str -> object) = Nome do passageiro;
5. `Sex`(str -> object) = Masculino e feminino. No dataset é 'male' ou 'female';
6. `Age`(float) = Idade do passageiro quando ocorreu o naufrágio;
7. `SibSp`(int) = Número de irmãos a bordo;
8. `Parch`(int) = Número de pais e/ou filhos a bordo;
9. `Ticket`(str) = Identificador unico. Código do ticket;
10. `Fare`(float) = Valor da passagem paga em dólares;
11. `Cabin`(str -> object) = Identificação da cabine;
12. `Embarked`(str -> object) = Local onde o passageiro embarcou.


#### Quais tipos de problemas que vamos resolver? Classificação, regressão ou  ranqueamento(por exemplo)?

R- O que nos foi repassado: criar um modelo de machine learning para que com base em uma entrada com as características de um indivíduo inferir se ele sobreviveu ou não ao naufrágio. Logo isso se trata de **modelo de classificação**.


#### Quais modelos vamos testar?

R- Utilizando modelos baseados em árvore e probabilístico: 

   - RandomForest;
   - Nayve Bayes;
   - LDA;
   
o interessante é que além de exibir 0 ou 1, podemos exibir a probabilidade das amostras, indicando a certeza do modelo.


#### Quais formas vamos testar esses modelos?

R- Como o problema é de classificação então, vamos usar estas métricas:

   - AUC;
   - Precisão;
   - Recall;

#### Quais são as métricas que vamos comparar para escolher o melhor modelo?

R- Utilizando o **AUC** e o **tempo de teste** são serão os principais para verificar a eficiência do modelo. Porque o tempo de teste é mais importante porque é o quanto que o **modelo demorará a inferir**. AUC é uma métrica geral para se dizer sobre se o modelo é melhor que o randômico.

## 1.1 Funções utilizadas

Funções utilizadas para plotar gráficos, testar os modelos e outras para tratar o dataframe para alguma visualização.

* **Dica**: Se esta sessão ficar muito grande crie arquivos `.py` cada um deles sendo módulos da forma que um servirá para plotar gráficos, tratar o dataframe, preparar os dados para alguma entreda de seu interesse, etc.*

```python
def plot_hist(dataframe, name_title):
    """
    Função para plotar um gráfico de barra exibindo o valor da barra logo acima no centro.
    
    Parâmetros
    ----------
    dataframe: Pandas.Dataframe
        Dataframe com duas colunas referente aos eixos x e y.
        
    name_title: str
        frase que será inserida como titulo do gráfico.
    
    """
    
    
    x_name = dataframe.columns[0]
    y_name = dataframe.columns[1]

    plt.figure(figsize=(8, 6))
    plots = sns.barplot(x=x_name, y=y_name, data=dataframe)

    for p in plots.patches:
        plots.annotate(format(p.get_height(), '.0f'), 
                       (p.get_x() + p.get_width() / 2., p.get_height()), 
                       ha = 'center', va = 'center', 
                       xytext = (0, 9), 
                       fontsize=15,
                       textcoords = 'offset points')


    plt.title(name_title,fontsize=15)    
    plt.ylabel("")
    plt.xlabel("")
    plt.xticks(fontsize=12)
    plt.ylim(0,800)
    plt.show()
    
def grid_search_model(model, parameteros, X, y, n_splits=10):
    """
    Realiza a procura dos melhores hiperparâmetros do modelo.
    
    Parâmetros
    ----------
    model: Model class
        Objeto de um modelo.
        
    parameteros: dict
        Conjunto de hiperparâmetros a se testar e gerar as combinações.
        
    X: pandas.DataFrame
        Dados de entrada. Padrão: colunas já definidas no dataset do titanic.
        
    y: pandas.DataFrame
        Rótulos das entradas. Padrão: coluna já definida no dataset do titanic. 
        
    n_splits: int
        Quantidade de divisões que kfold fará, padrão: 10.
        
    Return
    ------
    clf: GridSearchCV object
        Conjunto de modelos treinados e suas métricas.
        
    """


    # Números de divisões(splits) no dataset
    kf = KFold(n_splits=n_splits)

    # GridSearch
    clf = GridSearchCV(estimator = model,
                       param_grid = parameters,
                       cv = kf)
    clf.fit(X, y)

    return clf
    
def exibindo_resultados(clf, model, X, y):
    """
        Exibe as métricas de qualidade e performance do modelo para podermos julgarmos sua 
    eficiência em relação ao problema repassado.
     
    Parâmetros
    ----------
    clf: Objeto da classe GridSearchCV.
        Nele contêm todas as métricas dos modelos testados bem com os melhor conjunto 
        de hiperparâmetros. Se ele tivesse o tempo de treino e teste de cada um, apenas
        esse objeto serviria nesta função.
    
    model: Objeto instanciado da classe do Modelo.
        Objeto de um modelo de entrada.
                
    X: pandas.DataFrame
        Dados de entrada. Padrão: colunas já definidas no dataset do titanic.
        
    y: pandas.DataFrame
        Rótulos das entradas. Padrão: coluna já definida no dataset do titanic. 
    """
    
    # Primeiro - mostrando os resultados do melhor estimador e seu conjunto de hiperparâmetros
    print("- Melhor conjunto de hiperparâmetros:")
    for key, value in clf.best_params_.items():
        print(f" · {key}: {value}")

    print(f"- Métrica média deste conjunto - Kfold : {clf.best_score_}")

    ## Retreinando o modelo - observando com os melhores hiperparâmetros
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, shuffle=True)
    clf_best_estimator = model.set_params(**clf.best_params_)

    # Treinando e testando o modelo e observando seu tempo de treino
    start_time_train = datetime.now()
    clf_best_estimator.fit(X_train, y_train)
    total_time = (datetime.now() - start_time_train).total_seconds()
    y_pred = clf_best_estimator.predict(X_test)

    # Calculando o tempo médio de predição de resultados do modelo
    all_time_test = []
    for sample_test in X_test.values:
        start_time_test = datetime.now()
        clf_best_estimator.predict(sample_test.reshape(1, -1))
        all_time_test.append((datetime.now() - start_time_train).total_seconds())

    # Exibindo as estatisticas de tempo
    print(f'- Tempo total de treinamento: {total_time}s')
    print(f'- Tempo médio de testes: {round(np.mean(all_time_test),3)}s [±{round(np.var(all_time_test),3)}]')

    # Exibindo as métricas de classificação do modelo
    print("Métricas gerais sobre o modelo com os melhores hiperparâmetros")
    print(classification_report(y_test, y_pred))

    # Plotando a curva ROC e sua AUC
    print("Curva ROC e sua AUC")
    rf_probas = clf_best_estimator.predict_proba(X)
    fpr, tpr, thresholds = roc_curve(y, rf_probas[:, 1])
    plot_roc_curve(clf_best_estimator, X_test, y_test)
    plt.show()    
```

## 2.6 Completando os dados com *Expected Maximization*

O algoritmo possui duas etapas com o objetivo de imputar os dados a partir da função de máxima verossimilhança.  De forma iterativa; coloca valores aleatórios derivados de um estimador inicial para uma dada distribuição (geralmente normal). Primeiro passo, é chamado de expectativa (E), em que se completa os dados faltantes a partir de um distribuição, dado um estimador inicial; o segundo passo é a maximização (M) que atualiza o estimador e os dados faltantes encontrados no passo anterior da iteração. 

Para mais informações sobre como funciona,dá uma olhada no algoritmo dos autores, [aqui oh](https://rss.onlinelibrary.wiley.com/doi/10.1111/j.2517-6161.1977.tb01600.x)

```python
df_titanic.Age = imp.cs.em(df_titanic.Age.values.reshape(-1,1), loops=1000)
df_titanic.Age
```

## One-Enconding das colunas categoricas

```python
# Removendo a coluna 'Cabin'por ter muitos valores faltantes e valores faltantes da 'Embarked'
df_titanic = df_titanic.loc[:, df_titanic.columns != 'Cabin'].dropna()
df_titanic
```

Criando as colunas one-hot referentes as colunas categóricas.

```python
one_hot_pclass = []
one_hot_sex_class = []
one_hot_embarked = []

for classe in df_titanic.Pclass:
    if classe == 1:
        one_hot_pclass.append([1,0,0]) 
    elif classe == 2:
        one_hot_pclass.append([0,1,0])
    elif classe == 3:
        one_hot_pclass.append([0,0,1])
        
for classe in df_titanic.Sex:
    if classe == 'male':
        one_hot_sex_class.append([1,0]) 
    elif classe == 'female':
        one_hot_sex_class.append([0,1])

        
for classe in df_titanic.Embarked:
    if classe == "S":
        one_hot_embarked.append([1,0,0]) 
    elif classe == "C":
        one_hot_embarked.append([0,1,0])
    elif classe == "Q":
        one_hot_embarked.append([0,0,1])        
        
df_titanic[['Pclass1','Pclass2','Pclase3']]= np.array(one_hot_pclass)
df_titanic[['Embarked_S','Embarked_C','Embarked_Q']]= np.array(one_hot_embarked)
df_titanic[['Sex_male','Sex_female']] = np.array(one_hot_sex_class)
display(df_titanic.head())
print('\n\n')
print(df_titanic.columns)
```
